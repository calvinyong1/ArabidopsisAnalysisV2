```python
import cv2 
import nibabel as nib
import numpy as np
import os 

base_paths = ['Sistema_Placas/rpi102_2025-02-07_10-13']

all_paths = []

# This will search in all the cameras of a robot for the images to annotate
for path in base_paths:
    subfolders = os.listdir(path)
    for subfolder in subfolders:
        if os.path.isdir(os.path.join(path, subfolder)):
            all_paths.append(os.path.join(path, subfolder))

# Then this will store the .nii.gz segmentations in the same folders as the images
for path in all_paths:
    images = os.listdir(path)
    images = [img for img in images if img.endswith('.png')]
    
    for image in images:
        seg_path = os.path.join(path, "Segmentation", "Ensemble", image)
        
        seg = cv2.imread(seg_path, cv2.IMREAD_UNCHANGED)
        seg = seg.astype(np.uint8).T
        
        # save segmentation as .nii.gz
        seg_nii = nib.Nifti1Image(seg, affine=np.eye(4))
        seg_nii_path = os.path.join(path, image.replace('.png', '.nii.gz'))
        nib.save(seg_nii, seg_nii_path)
        print(f'Saved segmentation to {seg_nii_path}')
```
